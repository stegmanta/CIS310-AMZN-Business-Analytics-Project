```python
#Remove columns from data that have a constant/the same value
constant_columns=[]

for c in data.columns:
    #unique returns a list of unique values
    n = len(data[c].unique())
    if n == 1:
        constant_columns.append(c)
        
print(f"Constant columns: {constant_columns}")
```

```python
data.drop(columns=constant_columns, axis=1, inplace=True)
```
